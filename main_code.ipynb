### SHOWING TRAINING DATA AS IMAGES

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Load the training data from the pickle file
with open("test_data1.pkl", 'rb') as f:
    train_data = pickle.load(f)

n_rows = 3
n_cols = 3

fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(12, 12))

axs = axs.flatten()

for i, ax in enumerate(axs):
    if i >= len(train_data[0]):
        break
    matrix = train_data[0][i]
    ecc = train_data[1][i]
    theta = train_data[2][i]
    
    ax.imshow(matrix, cmap='gray', origin='lower', extent=[0, 100, 0, 100])
    ax.set_title(f'Ecc: {ecc:.2f}, Theta: {np.degrees(theta):.0f}')
    ax.axis('off')

for ax in axs[len(train_data[0]):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
