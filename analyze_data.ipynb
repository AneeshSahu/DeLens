```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Directory containing the .npy files
directory = r"lensingData/Data/train/Lensed"


# Get a list of .npy files in the directory
npy_files = [f for f in os.listdir(directory) if f.endswith('.npy')][:20]

# Sort the files to maintain order
npy_files.sort()

# Create subplots
n_files = len(npy_files)
n_cols = 4  # You can choose the number of columns for your subplot grid
n_rows = n_files // n_cols + (n_files % n_cols > 0)  # Calculate rows, make sure all plots fit

# Create a figure with subplots
fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, n_rows * 3))

# Flatten the axes array for easy iteration
axs = axs.flatten()

# Iterate through the list and plot each file
for idx, npy_file in enumerate(npy_files):
    # Load the .npy file
    data = np.load(os.path.join(directory, npy_file))
    
    # Plot the data
    axs[idx].imshow(data, cmap='viridis')
    axs[idx].set_title(npy_file)
    axs[idx].axis('off')  # Turn off axis lines and labels

# Hide any remaining unused subplots
for ax in axs[idx+1:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
